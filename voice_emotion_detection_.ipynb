Feature Extraction

```python
def extract_mfcc(filename):
     y, sr = librosa.load(filename, duration=3, offset=0.5)
     mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
     return mfcc
```

```python
extract_mfcc(df['speech'][0])
```

```python
X_mfcc = df['speech'].apply(lambda x: extract_mfcc(x))
```

```python
X_mfcc
```

```python
X = [x for x in X_mfcc]
X = np.array(X)
X.shape
```

```python
## input split
X = np.expand_dims(X, -1)
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y = enc.fit_transform(df[['label']])
```

```python
y = y.toarray()
```

```python
y.shape
```

Create the LSTM Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

model = Sequential([
    LSTM(256, return_sequences=False, input_shape=(40,1)),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Train the model
history = model.fit(X, y, validation_split=0.2, epochs=50, batch_size=64)
```

Plot the results

```python
epochs = list(range(50))
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.plot(epochs, acc, label='train accuracy')
plt.plot(epochs, val_acc, label='val accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs, loss, label='train loss')
plt.plot(epochs, val_loss, label='val loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
# === Necessary Imports ===
import base64
import json
import queue
import time # Added for sleep

import numpy as np
import librosa
import librosa.display # Explicitly import display
import matplotlib.pyplot as plt
from IPython.display import Javascript, display
from google.colab import output

# --- IMPORTANT: Load Your Model and Encoder ---
# You NEED to have uploaded your model and encoder files
# or mount your Google Drive where they are stored.
# Replace the paths below with the correct paths to your files.

# Example using TensorFlow/Keras and Pickle:
import tensorflow as tf
import pickle # Or import joblib if you saved with joblib

MODEL_PATH = '/content/emotion_detection_model.keras' # <-- REPLACE THIS PATH (e.g., '/content/drive/MyDrive/models/emotion_model.h5')
ENCODER_PATH = '/content/emotion_encoder.pkl' # <-- REPLACE THIS PATH (e.g., '/content/drive/MyDrive/models/encoder.pkl')

try:
    print(f"Loading model from: {MODEL_PATH}")
    model = tf.keras.models.load_model(MODEL_PATH)
    print(f"Loading encoder from: {ENCODER_PATH}")
    with open(ENCODER_PATH, 'rb') as f:
        enc = pickle.load(f) # Or use joblib.load(f) if applicable
    print("\n--- Model and Encoder loaded successfully ---")
    print(f"Model input shape expected: {model.input_shape}") # Print expected shape
    # You might want to print enc.classes_ here to see the loaded labels
    # print(f"Encoder classes: {enc.classes_}")

except Exception as e:
    print(f"\n--- ERROR loading model or encoder ---")
    print(f"Error details: {e}")
    print("\n--- PLEASE CHECK THE FOLLOWING: ---")
    print(f"1. Did you replace the placeholder paths correctly?")
    print(f"   - Model Path: '{MODEL_PATH}'")
    print(f"   - Encoder Path: '{ENCODER_PATH}'")
    print(f"2. Do the files actually exist at these paths?")
    print(f"   (Check spelling, Drive mount, upload status)")
    print(f"3. Are you using the correct loading function?")
    print(f"   (tf.keras.models.load_model for H5/SavedModel, pickle/joblib for encoder)")
    print("\n--- Stopping execution because essential files are missing. ---")
    # Stop the script if loading failed
    raise SystemExit("Model/Encoder loading failed.")
# --- End of Model and Encoder Loading ---


# === Audio Parameters ===
SAMPLE_RATE = 22050  # Sample rate expected by librosa/your model
CHANNELS = 1         # Mono audio
N_MFCC = 40          # Number of MFCC features to extract


# === Audio Queue ===
# This queue holds audio chunks received from the browser
audio_queue = queue.Queue()


# === JavaScript for Recording ===
# This code runs in your browser to capture audio
RECORD_JS = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader();
  reader.onloadend = function() {
    resolve(reader.result);
  }
  reader.readAsDataURL(blob);
});

var recordButton = null;
var stopButton = null;
var recorder = null;
var audioChunks = [];

// Function to initialize recorder and buttons
async function setupRecorder() {
    try {
        // Remove existing buttons if they exist
        const existingRecord = document.getElementById("recordButton");
        if (existingRecord) existingRecord.remove();
        const existingStop = document.getElementById("stopButton");
        if (existingStop) existingStop.remove();

        recordButton = document.createElement("button");
        recordButton.id = "recordButton";
        recordButton.innerHTML = "Start Recording";
        document.body.appendChild(recordButton);

        stopButton = document.createElement("button");
        stopButton.id = "stopButton";
        stopButton.innerHTML = "Stop Recording";
        stopButton.disabled = true;
        document.body.appendChild(stopButton);

        console.log("Requesting microphone access...");
        // IMPORTANT: This line triggers the browser's permission prompt
        const stream = await navigator.mediaDevices.getUserMedia({ audio: { sampleRate: SAMPLE_RATE, channels: CHANNELS } });
        console.log("Microphone access granted.");

        recorder = new MediaRecorder(stream);

        recorder.ondataavailable = event => {
          console.log("Audio data available...");
          audioChunks.push(event.data);
        };

        recorder.onstop = async () => {
          console.log("Recording stopped. Processing audio...");
          const audioBlob = new Blob(audioChunks);
          const arrayBuffer = await audioBlob.arrayBuffer();
          // Convert ArrayBuffer to Base64 string
          const audioAsBase64 = btoa(new Uint8Array(arrayBuffer).reduce((data, byte) => data + String.fromCharCode(byte), ''));
          // Send data back to Python kernel
          const json = JSON.stringify({ audio: audioAsBase64, sampleRate: SAMPLE_RATE, channels: CHANNELS });
          google.colab.kernel.invokeFunction('notebook.process_audio_chunk', [json], {});
          console.log("Audio data sent to Python.");
          // Reset for next recording
          audioChunks = [];
          recordButton.disabled = false;
          stopButton.disabled = true;
        };

        recordButton.onclick = () => {
          recorder.start();
          console.log("Recording started.");
          recordButton.disabled = true;
          stopButton.disabled = false;
        };

        stopButton.onclick = () => {
          recorder.stop();
          // Note: Actual processing happens in onstop handler
        };
         console.log("Recorder setup complete. Buttons added.");

    } catch (err) {
        console.error("Error accessing microphone or setting up recorder:", err);
        // Display error to the user in the Colab output
        google.colab.kernel.invokeFunction('notebook.show_js_error', [err.message], {});
    }
}

// Run the setup function
setupRecorder();

"""

# === Python Callback Function ===
# This function receives the audio data from JavaScript
def process_audio_chunk(json_string):
    """Decodes Base64 audio, converts to NumPy array, and adds to queue."""
    global audio_queue # Allow modification of the global queue
    try:
        data = json.loads(json_string)
        audio_base64 = data['audio']
        sample_rate = data['sampleRate']
        channels = data['channels']

        # Decode Base64
        audio_bytes = base64.b64decode(audio_base64)

        # Convert bytes to NumPy array
        # --- Potential Issue Point ---
        # MediaRecorder might output int16. If float32 below doesn't work well
        # (e.g., very quiet spectrogram, poor predictions), try int16:
        # audio_array = np.frombuffer(audio_bytes, dtype=np.int16)
        # audio_array = audio_array.astype(np.float32) / 32768.0 # Normalize
        # --- End Potential Issue Point ---
        audio_array = np.frombuffer(audio_bytes, dtype=np.float32)


        # Ensure it has the correct shape (samples, channels)
        # If reshape fails, audio might be empty or corrupted
        if audio_array.size == 0:
             print("Warning: Received empty audio chunk.")
             return
        audio_array = audio_array.reshape(-1, channels)

        print(f"Received audio chunk: {audio_array.shape[0]} samples, {audio_array.shape[1]} channels")
        audio_queue.put(audio_array)

    except Exception as e:
        print(f"Error processing audio chunk in Python callback: {e}")
        import traceback
        traceback.print_exc()

# Function to display JavaScript errors in Colab output
def show_js_error(error_message):
    print(f"\n--- JavaScript Error ---")
    print(f"Error reported from browser: {error_message}")
    print("This often means the browser could not access the microphone.")
    print("Please check browser permissions (microphone icon in address bar) and ensure a microphone is connected/enabled.")
    print("--- End JavaScript Error ---")

# Register the callback functions with Colab
output.register_callback('notebook.process_audio_chunk', process_audio_chunk)
output.register_callback('notebook.show_js_error', show_js_error)


# === Execute JavaScript and Start Main Loop ===

# Display the JavaScript, which should add buttons and request mic access
print("Executing JavaScript to add recording controls...")
print("!!! Browser will likely ask for Microphone Permission now. Please ALLOW it. !!!")
display(Javascript(RECORD_JS.replace('SAMPLE_RATE', str(SAMPLE_RATE)).replace('CHANNELS', str(CHANNELS))))

print("\nRecording controls should appear above (you might need to scroll up slightly).")
print("Click 'Start Recording', speak, and then 'Stop Recording'.")
print("Processing will occur after you click 'Stop Recording'. Waiting for audio...")

# Turn on interactive mode for Matplotlib for non-blocking plots
plt.ion()
fig, ax = plt.subplots(figsize=(8, 4)) # Create figure once

# Main loop to process audio from the queue
while True:
    if not audio_queue.empty():
        print("\nProcessing new audio chunk...")
        data = audio_queue.get() # Get audio data (NumPy array)

        # Ensure data is not empty before processing
        if data is None or data.size == 0:
            print("Skipping empty data chunk.")
            continue

        try:
            # --- Feature Extraction (MFCC) ---
            # Flatten if multi-channel (we expect mono, but safe practice)
            y = data.flatten()
            # Check duration - librosa needs non-empty audio
            if len(y) < 1:
                 print("Warning: Audio chunk too short after flattening.")
                 continue

            mfccs = librosa.feature.mfcc(y=y, sr=SAMPLE_RATE, n_mfcc=N_MFCC)
            # Average MFCCs over the time dimension for a single feature vector
            mfccs_processed = np.mean(mfccs.T, axis=0)

            # --- Reshape for Model Prediction ---
            # Add batch dimension: (N_MFCC,) -> (1, N_MFCC)
            mfccs_reshaped = np.expand_dims(mfccs_processed, axis=0)
            # Add channel dimension if needed: (1, N_MFCC) -> (1, N_MFCC, 1)
            # ** VERIFY: Does your model expect (batch, features) or (batch, features, 1)? **
            # Uncomment the next line ONLY if your model needs the trailing '1' dimension
            # mfccs_reshaped = np.expand_dims(mfccs_reshaped, axis=-1)

            # --- Predict Emotion ---
            print(f"Feeding features of shape {mfccs_reshaped.shape} to the model...")
            prediction = model.predict(mfccs_reshaped)
            # Get the predicted label index (highest probability)
            predicted_index = np.argmax(prediction, axis=1)[0]
            # Convert index to label using the loaded encoder
            predicted_label = enc.inverse_transform([predicted_index])[0]


            # --- Display Results ---
            print(f"\n---> Predicted Emotion: {predicted_label} <---")
            # print(f"(Raw prediction scores: {prediction[0]})") # Optional: show raw scores

            # --- Calculate and Display Spectrogram ---
            # Clear previous plot
            ax.clear()
            spectrogram = librosa.feature.melspectrogram(y=y, sr=SAMPLE_RATE)
            img = librosa.display.specshow(librosa.power_to_db(spectrogram, ref=np.max),
                                           y_axis='mel', x_axis='time', sr=SAMPLE_RATE, ax=ax)
            # Add colorbar only if it doesn't exist or needs recreation
            # Simple approach: always add colorbar, might cause slight layout shifts
            if 'colorbar' not in locals() or colorbar.ax.figure != fig:
                 try: # Remove old colorbar if fig changed
                      colorbar.remove()
                 except: pass
                 colorbar = fig.colorbar(img, ax=ax, format='%+2.0f dB')
            else: # Update existing colorbar
                 colorbar.update_normal(img)


            ax.set_title(f'Mel Spectrogram (Predicted: {predicted_label})')
            fig.canvas.draw()
            fig.canvas.flush_events() # Update the plot display
            print("Spectrogram updated.")


        except Exception as e:
            print(f"\n--- ERROR during audio processing or prediction ---")
            print(f"Error details: {e}")
            import traceback
            traceback.print_exc() # Print detailed traceback
            print("----------------------------------------------------")

    else:
        # Sleep briefly when the queue is empty to avoid busy-waiting
        time.sleep(0.1) # Sleep for 100ms

# Note: This loop runs indefinitely. You'll need to manually interrupt
# the kernel (Runtime -> Interrupt execution or Stop button) to stop it.
```
